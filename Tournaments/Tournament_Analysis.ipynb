```python
def pieces_sacrificed(final_pos):

    black_bishop_count = 0
    black_knight_count = 0
    black_rook_count = 0
    black_queen_count = 0
    black_king_count = 0
    black_pawn_count = 0
    black_sacrificed = 0

    white_bishop_count = 0
    white_knight_count = 0
    white_rook_count = 0
    white_queen_count = 0
    white_king_count = 0
    white_pawn_count = 0
    white_sacrificed = 0

    for piece in final_pos:
        if piece == 'b':
            black_bishop_count += 1
        if piece == 'n':
            black_knight_count +=1
        if piece == 'r':
            black_rook_count += 1
        if piece == 'q':
            black_queen_count +=1
        if piece == 'k':
            black_king_count += 1
        if piece == 'p':
            black_pawn_count += 1


        if piece == 'B':
            white_bishop_count += 1
        if piece == 'N':
            white_knight_count +=1
        if piece == 'R':
            white_rook_count += 1
        if piece == 'Q':
            white_queen_count +=1
        if piece == 'K':
            white_king_count += 1
        if piece == 'P':
            white_pawn_count += 1

    if black_bishop_count == 1:
        black_sacrificed += 1
    if black_bishop_count == 0:
        black_sacrificed += 2
    if black_knight_count == 1:
        black_sacrificed += 1
    if black_knight_count == 0:
        black_sacrificed += 2
    if black_rook_count == 1:
        black_sacrificed += 1
    if black_rook_count == 0:
        black_sacrificed += 2
    if black_queen_count == 0:
        black_sacrificed += 1
    if black_king_count == 0:
        black_sacrificed +=1
    
    if black_pawn_count == 7:
        black_sacrificed += 1
    if black_pawn_count == 6:
        black_sacrificed += 2
    if black_pawn_count == 5:
        black_sacrificed += 3
    if black_pawn_count == 4:
        black_sacrificed += 4
    if black_pawn_count == 3:
        black_sacrificed += 5
    if black_pawn_count == 2:
        black_sacrificed += 6
    if black_pawn_count == 1:
        black_sacrificed += 7
    if black_pawn_count == 0:
        black_sacrificed += 8
        
    if white_bishop_count == 1:
        white_sacrificed += 1
    if white_bishop_count == 0:
        white_sacrificed += 2
    if white_knight_count == 1:
        white_sacrificed += 1
    if white_knight_count == 0:
        white_sacrificed += 2
    if white_rook_count == 1:
        white_sacrificed += 1
    if white_rook_count == 0:
        white_sacrificed += 2
    if white_queen_count == 0:
        white_sacrificed += 1
    if white_king_count == 0:
        white_sacrificed +=1
    
    if white_pawn_count == 7:
        white_sacrificed += 1
    if white_pawn_count == 6:
        white_sacrificed += 2
    if white_pawn_count == 5:
        white_sacrificed += 3
    if white_pawn_count == 4:
        white_sacrificed += 4
    if white_pawn_count == 3:
        white_sacrificed += 5
    if white_pawn_count == 2:
        white_sacrificed += 6
    if white_pawn_count == 1:
        white_sacrificed += 7
    if white_pawn_count == 0:
        white_sacrificed += 8
    total_sac = white_sacrificed + black_sacrificed
    return white_sacrificed, black_sacrificed, total_sac
```
